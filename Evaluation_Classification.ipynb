```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = load_breast_cancer()
X, y = data.data, data.target
labels = data.target_names
feature_names = data.feature_names
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Add Gaussian noise to the data set
np.random.seed(42)  # For reproducibility
noise_factor = 0.5 # Adjust this to control the amount of noise
X_noisy = X_scaled + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=X.shape)

# Load the original and noisy data sets into a DataFrame for comparison and visualization
df = pd.DataFrame(X_scaled, columns=feature_names)
df_noisy = pd.DataFrame(X_noisy, columns=feature_names)
```

### Exercise 1. Split the data, and fit the KNN and SVM models to the noisy training data


```python
# Split the data set into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_noisy, y, test_size=0.3, random_state=42)
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the models
knn = KNeighborsClassifier(n_neighbors=5)
svm = SVC(kernel='linear', C=1, random_state=42)

# Fit the models to the training data
knn.fit(X_train, y_train)
svm.fit(X_train, y_train)
```

### Evaluate the models
#### Predict on the test set


```python
y_pred_knn = knn.predict(X_test)
y_pred_svm = svm.predict(X_test)
```

#### Print the accuracy scores and classification reports for both models


```python
print(f"KNN Testing Accuracy: {accuracy_score(y_test, y_pred_knn):.3f}")
print(f"SVM Testing Accuracy: {accuracy_score(y_test, y_pred_svm):.3f}")

print("\nKNN Testing Data Classification Report:")
print(classification_report(y_test, y_pred_knn))

print("\nSVM Testing Data Classification Report:")
print(classification_report(y_test, y_pred_svm))
```

#### Plot the confusion matrices


```python
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)
conf_matrix_svm = confusion_matrix(y_test, y_pred_svm)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.heatmap(conf_matrix_knn, annot=True, cmap='Blues', fmt='d', ax=axes[0],
            xticklabels=labels, yticklabels=labels)

axes[0].set_title('KNN Testing Confusion Matrix')
axes[0].set_xlabel('Predicted')
axes[0].set_ylabel('Actual')

sns.heatmap(conf_matrix_svm, annot=True, cmap='Blues', fmt='d', ax=axes[1],
            xticklabels=labels, yticklabels=labels)
axes[1].set_title('SVM Testing Confusion Matrix')
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')

plt.tight_layout()
plt.show()
```

```python
print("Summary of Confusion Matrices:")

# KNN Confusion Matrix Summary
tn_knn, fp_knn, fn_knn, tp_knn = conf_matrix_knn.ravel()
print("\nKNN Model:")
print(f"True Positives (TP): {tp_knn}")
print(f"True Negatives (TN): {tn_knn}")
print(f"False Positives (FP): {fp_knn}")
print(f"False Negatives (FN): {fn_knn}")

# SVM Confusion Matrix Summary
tn_svm, fp_svm, fn_svm, tp_svm = conf_matrix_svm.ravel()
print("\nSVM Model:")
print(f"True Positives (TP): {tp_svm}")
print(f"True Negatives (TN): {tn_svm}")
print(f"False Positives (FP): {fp_svm}")
print(f"False Negatives (FN): {fn_svm}")
```

Here are the definitions of Accuracy, Precision, and Recall based on the confusion matrix:

*   **Accuracy:** The proportion of correctly classified instances out of the total instances. It is calculated as (TP + TN) / (TP + TN + FP + FN).
*   **Precision:** The ability of the model to correctly identify only the relevant instances (e.g., malignant cases). It is calculated as TP / (TP + FP). High precision means a low false positive rate.
*   **Recall (Sensitivity):** The ability of the model to find all the relevant instances (e.g., all malignant cases). It is calculated as TP / (TP + FN). High recall means a low false negative rate.

K-Nearest Neighbors (KNN) is a simple algorithm that classifies a data point based on the majority class of its 'k' nearest neighbors in the training data. It's a non-parametric and lazy learning algorithm, meaning it doesn't make assumptions about the underlying data distribution and memorizes the training data.

Support Vector Machine (SVM) is a powerful algorithm that finds the optimal hyperplane to separate different classes in the data. It aims to maximize the margin between the classes, which can lead to better generalization on unseen data. SVM can use different kernels (like linear, polynomial, or radial basis function) to handle non-linearly separable data.

Based on the results from our analysis:

Accuracy: The SVM model had a higher testing accuracy (0.971) compared to the KNN model (0.936).
Classification Report: The SVM model shows slightly better precision, recall, and f1-score for both classes (malignant and benign) than the KNN model.
Confusion Matrices: The SVM model had fewer false positives (2 vs 7) and false negatives (3 vs 4) than the KNN model. False positives are particularly important in this context as they represent cases where the model incorrectly predicts a malignant tumor (when it's benign), which could lead to unnecessary stress and treatment.
Therefore, based on these metrics, the SVM model performed better than the KNN model on this noisy dataset for this particular split of the data and chosen hyperparameters.

Exercise 6. Comparing training and testing accuracies for both models
What can you say about the accuracy of the two models on the training and test data sets?

What do these results possibly indicate?

Ideally the accuracy of a model would be almost the same on the training and testing data sets.

It would be unusual for the accuracy to be higher on the test set and this might occur due to chance or some sort of data leakage. For example, here we have normalized all of the data rather than fitting StandardScaler to the training data and only then applying it to the train and test sets separately. We'll revisit this and other pitfalls in another lab.

When the accuracy is substantially higher on the training data than on the testing data, the model is likely memorizing details in the training data that don't generalize to the unseen data - the model is overfitting to the training data.


| Model | Phase |  Accuracy |
| ------------  | -------- | --------- |
| KNN  | Train  | 95.5% |
| KNN  | Test   | 93.6% |
| SVM  | Train  | 97.2% |
| SVM  | Test   | 97.1% |

For the SVM model, the training and testing accuracies are essentially the same at about 97%. This is ideal - the SVM model is likely not overfit.
For the KNN model, however, the training accuracy is about 2% higher that the test accuracy, indicating there might be some overfitting.

In summary, the SVM model is both more convincing and has a higher accuracy than the KNN model.
